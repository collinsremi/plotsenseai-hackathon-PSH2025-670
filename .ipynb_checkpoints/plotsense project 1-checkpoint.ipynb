Air Quality Index (AQI) Labeling
--------------------------------
After going through our data, we transform raw air pollution data (PM2.5, PM10, NO2, SO2, CO, O3)
into an Air Quality Index (AQI) score and category.

**🔑 Why?**
The AQI is a standardized measure used worldwide to describe how polluted the air is 
and what associated health effects might be. It converts complex pollutant concentrations
into an easy-to-understand number and category (Good, Moderate, Unhealthy, etc.).

**⚙️ How?**
Each pollutant has official breakpoints defined by environmental agencies (e.g., US EPA).
We define functions for each pollutant that map its concentration → AQI sub-index.
The overall AQI for a location is the **maximum sub-index** (the worst pollutant dominates).

**🎯 Goal:**
- Add new AQI columns (per pollutant + overall AQI).
- Create an "AQI_Category" column = classification labels for ML models.

**__________________________________________________________________________________________________________________________________________**

```python
## -------------------------------------------------------------------
# AQI Functions
# Each function converts pollutant concentration → AQI sub-index
# based on EPA breakpoints. These are simplified but capture
# the essence of AQI calculation.
# -------------------------------------------------------------------


def aqi_pm25(pm25):
    if pm25 <= 12: return 50
    elif pm25 <= 35.4: return 100
    elif pm25 <= 55.4: return 150
    elif pm25 <= 150.4: return 200
    elif pm25 <= 250.4: return 300
    else: return 500

def aqi_pm25(pm25):
    if pm25 <= 12: return 50
    elif pm25 <= 35.4: return 100
    elif pm25 <= 55.4: return 150
    elif pm25 <= 150.4: return 200
    elif pm25 <= 250.4: return 300
    else: return 500

def aqi_pm10(pm10):
    if pm10 <= 54: return 50
    elif pm10 <= 154: return 100
    elif pm10 <= 254: return 150
    elif pm10 <= 354: return 200
    elif pm10 <= 424: return 300
    else: return 500

def aqi_no2(no2):
    if no2 <= 53: return 50
    elif no2 <= 100: return 100
    elif no2 <= 360: return 150
    elif no2 <= 649: return 200
    elif no2 <= 1249: return 300
    else: return 500

def aqi_so2(so2):
    if so2 <= 35: return 50
    elif so2 <= 75: return 100
    elif so2 <= 185: return 150
    elif so2 <= 304: return 200
    elif so2 <= 604: return 300
    else: return 500

def aqi_co(co):
    if co <= 4.4: return 50
    elif co <= 9.4: return 100
    elif co <= 12.4: return 150
    elif co <= 15.4: return 200
    elif co <= 30.4: return 300
    else: return 500

def aqi_o3(o3):
    if o3 <= 54: return 50
    elif o3 <= 70: return 100
    elif o3 <= 85: return 150
    elif o3 <= 105: return 200
    elif o3 <= 200: return 300
    else: return 500

        
# Apply functions
df["AQI_PM25"] = df["PM2.5"].apply(aqi_pm25)
df["AQI_PM10"] = df["PM10"].apply(aqi_pm10)
df["AQI_NO2"]  = df["NO2"].apply(aqi_no2)
df["AQI_SO2"]  = df["SO2"].apply(aqi_so2)
df["AQI_CO"]   = df["CO"].apply(aqi_co)
df["AQI_O3"]   = df["O3"].apply(aqi_o3)

# Take the maximum AQI across pollutants (worst case)
df["AQI"] = df[["AQI_PM25","AQI_PM10","AQI_NO2","AQI_SO2","AQI_CO","AQI_O3"]].max(axis=1)

# Map AQI to categories
def aqi_category(aqi):
    if aqi <= 50: return "Good"
    elif aqi <= 100: return "Moderate"
    elif aqi <= 150: return "Unhealthy-Sensitive"
    elif aqi <= 200: return "Unhealthy"
    elif aqi <= 300: return "Very Unhealthy"
    else: return "Hazardous"

df["AQI_Category"] = df["AQI"].apply(aqi_category)

```
